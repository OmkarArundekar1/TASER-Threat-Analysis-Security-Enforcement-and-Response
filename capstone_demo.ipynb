```python
import torch
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
import numpy as np

data = np.load("ml_pipeline/window_features.npy")  # fix path if needed
data = torch.tensor(data, dtype=torch.float32)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

import torch
data_scaled = torch.tensor(data_scaled, dtype=torch.float32)
```

```python
model = Autoencoder(input_dim=data_scaled.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = torch.nn.MSELoss()

epochs = 10

for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(data_scaled)
    loss = criterion(outputs, data_scaled)
    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch+1}, Loss: {loss.item():.6f}")
```
